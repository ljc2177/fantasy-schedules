```python
import pandas as pd
import random
import math
```

All non-conference teams will be played once, and the rest of the games will be made up of conference teams. If the total conference games is divisible by the number of conference teams minus one (how many teams each team would need to play within their conference), then each team will play the teams in their conference however many times that is. If it is not divisible, All confernece teams will be played a minimum number of games, followed by however many games need to be played to hit the total number of games and each team will not be played the same number of times. For best results, make sure the number of weeks after all non-conference games are played is divisible by conference teams minus one.

```python
weeks = 19

# team conference lists
conference1 = ["Team 1","Team 2","Team 3","Team 4","Team 5","Team 6","Team 7"] # conferences must be equal
conference2 = ["Team 8","Team 9","Team 10","Team 11","Team 12","Team 13","Team 14"] # conferences must be equal

# time between teams playing each other a second time
buffer = 5
```

```python
# define function to determine schedule using total number of weeks, 

def season_schedule(week_count,conference1_teams,conference2_teams,game_buffer):
    # define fillable sets for team schedules, weekly schedules, and potential opponents
    all_teams = conference1_teams + conference2_teams
    schedule = {w: [] for w in range(1,week_count+1)}
    team_schedule = {t: [] for t in all_teams}
    potential_opponents = {t: [] for t in all_teams}

    # calculate total number of conference rounds
    non_conference_games = len(conference1_teams)
    conference_games = week_count - non_conference_games
    total_conference_rounds = math.ceil(conference_games / (len(conference1_teams)-1))

    # determine potential opponents for each team

    for team in all_teams:
        potential_opponents[team] = all_teams
        for round in range(0,int(total_conference_rounds)-1):
            if team in conference1_teams:
                potential_opponents[team] = potential_opponents[team] + conference1_teams # add conference1_all_teams all_teams twice to account for Xx games for in-conference all_teams
                potential_opponents[team] = [i for i in potential_opponents[team] if i != team]
                # print(potential_opponents[team]) # uncomment if you want to check potential opponents calcs
            else:
                potential_opponents[team] = potential_opponents[team] + conference2_teams # add conference2_all_teams all_teams twice to account for Xx games for in-conference all_teams
                potential_opponents[team] = [i for i in potential_opponents[team] if i != team]
                # print(potential_opponents[team]) # uncomment if you want to check potential opponents calcs

    # determine schedule

    games_per_week = len(all_teams)/2
    for week in range(1,week_count+1):
        # print("week "+str(week)) # uncomment to track progress if there are errors
        for game in range(0,int(games_per_week)):
            # print("game "+str(game)) # uncomment to track progress if there are errors
            team1 = random.choice(all_teams)
            while team1 in schedule[week]:
                team1 = random.choice(all_teams)
            schedule[week] = schedule[week] + [team1]
            potential_opponent = potential_opponents[team1]
            potential_opponent = [i for i in potential_opponent if i not in schedule[week]]
            start_date = week - game_buffer
            if start_date < 0:
                start_date = 0
            else:
                start_date = start_date
            potential_opponent = [i for i in potential_opponent if i not in team_schedule[team1][start_date:week]]
            team2 = random.choice(potential_opponent)
            schedule[week] = schedule[week] + [team2]
            team_schedule[team1] = team_schedule[team1] + [team2]
            team_schedule[team2] = team_schedule[team2] + [team1]
            potential_opponents[team1].remove(team2)
            potential_opponents[team2].remove(team1)

    return potential_opponents, team_schedule
```
